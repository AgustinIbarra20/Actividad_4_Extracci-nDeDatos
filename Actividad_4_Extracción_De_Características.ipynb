#LIMPIEZA DEL ARCHIVO DE VALORES NULOS Y OUTLIERS

```python
#carga desde un archivo .xlsx sin indices
Micro_Retailer=pd.read_excel('microretailer_mit_lift_lab.xlsx')
```

```python
Micro_Retailer.info()
```

```python
#ajustar maximo de columnas
pd.options.display.max_columns= None
```

```python
#Imprimo los primeros 5 registros
Micro_Retailer.head()
```

```python
#Filtro de las 20 columnas por categoricas cualitativas y categoricas cuantitativas
CAT_CUANTITATIVAS=Micro_Retailer.iloc[ :,[13,16,78]]
CAT_CUALITATIVAS=Micro_Retailer.iloc[ :,[3,9,36,47,48,53,54,67,68,75,80,82,86,91,92,96,99]]
```

```python
#Corroboramos valores nulos
valores_nulos=CAT_CUANT.isnull().sum()
valores_nulos
```

```python
CAT_CUANT=CAT_CUANT.fillna(method='ffill')
```

```python
CAT_CUANT=CAT_CUANT.fillna(method='bfill')
```

```python
#Corroboramos valores nulos
valores_nulos=CAT_CUANT.isnull().sum()
valores_nulos
```

```python
y=CAT_CUANT
Limite_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print("Limite superior permitido: \n", Limite_Superior)
print("Limite inferior permitido: \n", Limite_Inferior)
```

```python
#Encontramos Outliers del data frame
outliers = CAT_CUANT[(y>Limite_Superior) | (y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
CAT_CUANT = CAT_CUANT[(y<=Limite_Superior) & (y>=Limite_Inferior)]
CAT_CUANT
```

```python
#Corroboramos valores nulos
valores_nulos=CAT_CUANT.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del data frame con "mean"
data_clean_CAT_CUANT=CAT_CUANT.copy()
data_clean_CAT_CUANT=data_clean_CAT_CUANT.fillna(round(CAT_CUANT.mean(), 1))
data_clean_CAT_CUANT
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean_CAT_CUANT.isnull().sum()
valores_nulos
```

```python
#Corroboramos valores nulos
valores_nulos=CAT_CUAL.isnull().sum()
valores_nulos
```

```python
data_clean_CAT_CUAL=CAT_CUAL.copy()
data_clean_CAT_CUAL=data_clean_CAT_CUAL.fillna('DESCONOCIDO')
data_clean_CAT_CUAL
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean_CAT_CUAL.isnull().sum()
valores_nulos
```

```python
#Unimos los data frames para crear un Micro_Retailer ya limpio y sin outliers
Micro_Retailer = pd.concat([data_clean_CAT_CUAL, data_clean_CAT_CUANT], axis = 1)
Micro_Retailer
```
